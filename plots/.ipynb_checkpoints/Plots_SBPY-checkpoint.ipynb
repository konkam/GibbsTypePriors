## SB truncation

```julia
R"StickBreakingPY <- function(alpha,sigma, N) {
  a<- rep(1-sigma,N-1)
  b<- alpha+ sigma*(1:N-1)
  V <- rbeta(N-1 , a, b)
  p    <- vector('numeric',length=N)
  p[1] <- V[1]
  for(l in 2:(N - 1))p[l] <- prod(1 - V[1:(l - 1)])*V[l]
  p[N] <- prod(1 - V)   
  p
  return(p)
}

Prior_on_K_SB<- function(alpha, sigma, N_s,N_tr, runs=10^4 ){
  array_nc<-c()
  i=1
  while (i<=runs){
    weights_PY<- StickBreakingPY(alpha,sigma,N_tr)
    c <- sample(1:N_tr,size=N_s, replace=TRUE, prob=weights_PY )
    n_c<- length(unique(c))
    array_nc[i]<- n_c
    i=i+1
  }
  p_k = tibble(k=as.numeric(names(table(array_nc))), 
               p_k=as.vector(table(array_nc))/sum(table(array_nc)))
  p_zeros= tibble(k=(1:N_tr)[!1:N_tr%in%p_k$k], 
                  p_k=rep(0,length((1:N_tr)[!1:N_tr%in%p_k$k])))
  return(rbind(p_k, p_zeros))
}


p = tibble(k=1:50,
           pDP= Prior_on_K_SB(1, 0,N_s=100,N_tr=50, runs=10^5)$p_k, 
           pPY1 = Prior_on_K_SB(1, 0.25,N_s=100,N_tr=50, runs=10^5)$p_k,
           pPY2 = Prior_on_K_SB(1, 0.5,N_s=100,N_tr=50, runs=10^5)$p_k,
           pPY3 = Prior_on_K_SB(1, 0.75,N_s=100,N_tr=50, runs=10^5)$p_k,
           pPY4 = $(Pkn_2PD.(1:50, 100,  1.0, 0.25)))%>%
  gather(Process_type, distribution, pDP:pPY4)%>%
 ggplot(aes( x = k, y = distribution, color=Process_type )) +
theme_bw() +
geom_point() +scale_color_viridis(discrete=TRUE) +
geom_line()
p"
```

```julia
R"p = tibble(k=1:50,
           pPY0= (c(Prior_on_K_SB(0.98, 0.4,N_s=111,N_tr=20, runs=10^6)$p_k,rep(0,30))), 
           pPY1 = (c(Prior_on_K_SB(0.98, 0.4,N_s=111,N_tr=37, runs=10^6)$p_k,rep(0,13))),
           pPY2 = (Prior_on_K_SB(0.98, 0.4,N_s=111,N_tr=50, runs=10^6)$p_k),
           pPY3 = (Prior_on_K_SB(0.98, 0.4,N_s=111,N_tr=100, runs=10^6)$p_k)[1:50],
           pPY4 = $(Pkn_2PD.(1:50, 111, 0.98, 0.4)))%>%
  gather(Truncation, density, pPY0:pPY4)%>%
 ggplot(aes( x = k, y = density, color=Truncation )) +
theme_bw() +
geom_point() +
scale_color_manual(labels = c('N=20','N=37','N=50','N=100','no truncation'),
values = c( '#440154FF' ,'#CD4071FF','#3B528BFF',  '#5DC863FF', '#FDE725FF'))+ 
geom_line()
p"
R"pdf('Priors_PY_SB20_04_16.pdf')
plot(p)
dev.off()"
R"p"

```

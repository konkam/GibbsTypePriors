Reccursive approximation

```julia
function Pkn_NGG_approx_reccursive(n, start_ind, end_ind,β, σ, f, weight)
        Axnk = Array{arb}(undef, end_ind-start_ind)
        if length(Axnk)==0 
          return weight
        end
        Axnk[1:end_ind-start_ind] = f.(n, start_ind:(end_ind-1), β, σ)
        Sum_xn= RR(1)
        for i in (1:(end_ind-start_ind))
             Sum_xn= Sum_xn + exp(sum(Axnk[1:i]))
        end
        P1n= exp(log(weight)- log(Sum_xn))
        return P1n
end

function Pkn_NGG_approx_parts(n, start,β, σ)
        P1n = Array{arb}(undef, n)
        #int_n_half::Int64 = floor(n/2)
        P1n[start:n] = Pkn_NGG_arb.(start:n, n,  β, σ)
        left_weight = 1 - sum(P1n[start:n])
        for ki in (1:(start -1))
            P1n[ki] = Pkn_NGG_approx_reccursive(n, ki, start-1,β, σ, logxk, left_weight)
            left_weight = left_weight - P1n[ki]
        end
        return P1n
end

function approximation_partial_rec_df(β,σ,N)
    df= DataFrame(Pkn_exact = Pkn_NGG_raw.(1:N, N, β, σ),
                  Pkn_approx1 = Pkn_NGG_approx_part(N,  β, σ,Int64(floor(N/2))),
                  Pkn_approx2 = Pkn_NGG_approx_part(N,  β, σ,Int64(floor(N/4))),
                  Pkn_approx_rec =  Pkn_NGG_approx_parts(N,Int64(floor(N/4)), β, σ))
       return df
end
```

```julia
function compute_error_abs(Pk, Pk_exact)
  Error = Array{arb}(undef, length(Pk))
    for j in (1:(length(Pk)))
        Error[j] = abs(Pk[j] - Pk_exact[j]) //  Pk_exact[j]
     end
    abs_error= convert(Array{Float64},Error)
    return abs_error
end

function compute_error_rel(Pk, Pk_exact)
    Error = Array{arb}(undef, length(X))
    for j in (1:(length(Pk)))
        Error[j] = (Pk[j] - Pk_exact[j])
     end
     rel_error= convert(Array{Float64},Error)
    return rel_error
end

```

# Testing the approximation

1. Full approximation [Full] 
2. Approximation using some exact values $p_k$, $k=1,\cdots n$ [Partial] + substract last weight
3. Approximation using some exact values $p_k$, $k=1,\cdots n$ [Partial] + start from p1 -> possible jump
3. The reccursive approximation [stick]

```julia
## change added Pkn_NGG_approx_full
function Pkn_NGG_approxim(n, β, σ, f,ind_start, ind_end, stick)
        Axnk = Array{arb}(undef, ind_end-ind_start )
        Axnk[1:(ind_end-ind_start)] = f.(n, ind_start:(ind_end-1), β, σ)
        Sum_xn= RR(1)
        if length(Axnk)==1 
              return (1- stick)
        else 
            for i in (1:(ind_end-ind_start))
                 Sum_xn= Sum_xn + exp(sum(Axnk[1:i]))
            end
        end
        return  exp(log(stick)- log(Sum_xn))
end


Pkn_NGG_full_approxima(n, β, σ, f,ind_start, ind_end, stick) = convert(Float64, Pkn_NGG_approxim(n,  β, σ, f, ind_start, ind_end, stick))

```

```julia
function Pkn_NGG_approx_partial2(k, n, β, σ, start_k_ind, start_Pkn_val,f)
    if k==start_k_ind
        return start_Pkn_val
    else
        Pkn= Pkn_NGG_arb(start_k_ind, n,  β, σ)
        stick =sum(Pkn_NGG_arb.(start_k_ind:n, n,  β, σ)) 
        for ki in 1:k
                Pkn = Pkn_NGG_approxim(n, β, σ, f,ki,start_k_ind,stick)
                 stick= stick + Pkn
             end
         end
    return Pkn
end
```

```julia
function Pkn_NGG_AP(k, n, β, σ, start_k_ind,f)
    if k >= start_k_ind
        return Pkn_NGG_arb(k, n,  β, σ)
    else
        Pkn= RR(1)
        stick =sum(Pkn_NGG_arb.(start_k_ind:n, n,  β, σ)) 
        println(stick)
        for ki in 1:k
                Pkn = Pkn_NGG_approxim(n, β, σ, f,ki,start_k_ind,stick)
                 stick= stick + Pkn
             end
         end
    return Pkn
end
```

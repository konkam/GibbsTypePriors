```julia
function EV_of_number_of_clusters_NGG(n,vec)
    pk = Pkn_NGG.(1:n, n,vec[1], vec[2]) 
    E =  pk|> ar -> map(*, ar, 1:n) |> sum
    x = ((1:n).-E).^2
    V = pk |> ar -> map(*, ar, x) |> sum
    return E, V
end
```

```julia
function EV_of_number_of_clusters_NGG_multinomial(n,n_tr, vec)
    pk = Pkn_NGG.(1:n,n_tr,vec[1], vec[2]) 
    E =  pk|> ar -> map(*, ar, 1:n) |> sum
    x = ((1:n).-E).^2
    V = pk |> ar -> map(*, ar, x) |> sum
    return E, V
end
```

```julia
X= first.(EV_NGG)
Y = last.(EV_NGG)
DF_ev_NGGM = DataFrame(Exp = first.(EV_NGG),Var =last.(EV_NGG), beta = first.(grid_borders), sigma = last.(grid_borders))
DF_ev_NGGM.std = sqrt.(DF_ev_NGGM.Var)
DF_ev_NGGM
```
